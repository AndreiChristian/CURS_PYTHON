# Principiile SOLID: Un Ghid pentru Dezvoltarea Software-ului

## Ce sunt Principiile SOLID?

SOLID este un acronim pentru:

- **S**: Single Responsibility Principle (SRP) - Principiul Unicătății Responsabilității
- **O**: Open-Closed Principle (OCP) - Principiul Deschiderii-Închiderii
- **L**: Liskov Substitution Principle (LSP) - Principiul Substituției Liskov
- **I**: Interface Segregation Principle (ISP) - Principiul Segregării Interfețelor
- **D**: Dependency Inversion Principle (DIP) - Principiul Inversării Dependențelor

### Single Responsibility Principle (SRP)

#### Ce este?
O clasă ar trebui să aibă un singur motiv pentru a se schimba.

#### De ce contează?
Facilitează înțelegerea și întreținerea codului.

#### Pro:
- Ușor de depanat și testat.
- Ușor de citit și înțeles.

#### Contra:
- S-ar putea să te trezești cu multe clase mici, ceea ce poate fi coplesitor.

---

### Open-Closed Principle (OCP)

#### Ce este?
Entitățile software ar trebui să fie deschise pentru extensie, dar închise pentru modificare.

#### De ce contează?
Permite adăugarea de noi funcționalități fără a strica codul existent.

#### Pro:
- Flexibilitate și cuplare redusă.
- Mai ușor de adăugat funcționalități fără a rescrie codul existent.

#### Contra:
- Poate necesita mai multă muncă inițială pentru a face clasele extensibile.

---

### Liskov Substitution Principle (LSP)

#### Ce este?
Subtipurile trebuie să poată fi înlocuite cu tipurile lor de bază.

#### De ce contează?
Asigură faptul că utilizarea subclaselor nu va cauza comportament neașteptat în aplicație.

#### Pro:
- Face codul mai robust.
- Ajută la menținerea ierarhiilor de clase.

#### Contra:
- Restrânge modul în care îți proiectezi subclasele.

---

### Interface Segregation Principle (ISP)

#### Ce este?
Clienții nu ar trebui să fie forțați să depindă de interfețe pe care nu le folosesc.

#### De ce contează?
Minimizează suprafața pe care clienții trebuie să o cunoască, facilitând implementarea.

#### Pro:
- Mai ușor de implementat și întreținut pentru clasele client.
- Reduce cuplarea între clase.

#### Contra:
- Poate duce la fragmentarea în numeroase interfețe mici.

---

### Dependency Inversion Principle (DIP)

#### Ce este?
Depinde de abstracții, nu de implementări concrete.

#### De ce contează?
Decuplează codul, făcându-l mai ușor de extins și întreținut.

#### Pro:
- Codul este mai ușor de extins și întreținut.
- Cuplare redusă între clasele de nivel înalt și cele de nivel jos.

#### Contra:
- Complexitatea poate crește din cauza stratului de abstractizare.
