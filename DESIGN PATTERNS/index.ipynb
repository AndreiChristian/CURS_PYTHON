# DESIGN PATTERNS

### Creational Patterns

Acestea se ocupă de mecanismele de creare a obiectelor. Le-ai folosi atunci când vrei ca sistemul să fie independent de modul în care obiectele sale sunt create, compuse și reprezentate.

Șabloanele de creare te ajută să izolezi logica de creare a obiectelor de logica principală a aplicației tale. Asta înseamnă că poți schimba cum sunt create obiectele fără să zguduie tot sistemul.

- **Singleton**

- **Factory Method**

- **Abstract Factory**

- **Builder**

- **Prototype**

### Structural Patterns
These help ensure that when one part of a system changes, the entire structure of the system doesn't need to do the same. They provide a manner to define relationships between classes or objects.

- **Adapter**

- **Bridge**

- **Composite**
  
- **Decorator**

- **Facade**

- **Flyweight**

- **Proxy**

### Behavioral Patterns
These are all about object collaboration. Behavioral patterns help ensure that you do not have classes and objects that just stand there doing nothing.

- **Command**

- **Chain of Responsibility**

- **Interpreter**
  
- **Iterator**

- **Mediator**

- **Memento**

- **Observer**

- **State**

- **Strategy**

- **Template Method**

- **Visitor**
