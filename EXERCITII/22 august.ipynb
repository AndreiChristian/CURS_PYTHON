</ol>

<h2 style="color:gold">5. Programare Orientata pe Obiect</h2>

<ol>
    <li>Creeaza o clasa numita `Automobil` care sa aiba atributele: `marca` si `anul`.
        <span style="color:#ccc">Hint: Foloseste metoda constructor `__init__` pentru a initializa atributele.</span>
    </li>

   <li>Adauga o metoda in clasa `Automobil` care sa afiseze marca si anul automobilului.
        <span style="color:#ccc">Hint: Metodele sunt definite similar functiilor, dar in interiorul unei clase.</span>
    </li>
    
   <li>Creeaza un obiect din clasa `Automobil` si acceseaza atributele sale.
        <span style="color:#ccc">Hint: Obiectele sunt instante ale unei clase.</span>
    </li>
    
   <li>Modifica valoarea unui atribut al obiectului creat.
        <span style="color:#ccc">Hint: Poti accesa si modifica atributele unui obiect folosind notatia cu punct (.)</span>
    </li>

   <li>Creeaza o subclasa `MasinaElectrica` care mosteneste clasa `Automobil` si adauga un atribut nou: `autonomie`.
        <span style="color:#ccc">Hint: Mostenirea se realizeaza prin specificarea clasei parinte in paranteze la definirea noii clase.</span>
    </li>
    
   <li>Adauga o metoda in `MasinaElectrica` care sa afiseze autonomia.
        <span style="color:#ccc">Hint: Poti adauga metode noi intr-o subclasa similar clasei parinte.</span>

 <li>Modifica metoda de afisare din `Automobil` in `MasinaElectrica` astfel incat sa afiseze si autonomia.

<li>Creeaza o metoda statica in `Automobil` care sa verifice daca un an dat este an bisect.
        <span style="color:#ccc">Hint: Metodele statice nu primesc argumentul `self` si sunt decorate cu `@staticmethod`.</span>

        <span style="color:#ccc">Hint: Metodele de clasa primesc ca prim argument `cls` si sunt decorate cu `@classmethod`. Pentru a numara obiectele, ai nevoie de un atribut de clasa (ex. `numar_automobile`).</span>
    </li>
    
<li>Encapsuleaza atributele clasei `Automobil`, asigurandu-te ca anul nu poate fi setat la o valoare mai mica decat 1900.
        <span style="color:#ccc">Hint: Foloseste proprietati (getters si setters) pentru a controla accesul la atribute.</span>
    </li>
</ol>
